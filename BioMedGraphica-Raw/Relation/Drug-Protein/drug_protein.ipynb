### DrugBank

```python
# Download Link: https://go.drugbank.com/releases/5-1-13/downloads/target-all-uniprot-links
# Download Date: 2025-03-21
# Download Version: 2025-01-03

import pandas as pd

df_drugbank_drug_protein = pd.read_csv('uniprot links.csv')
drugbank_drug_protein = df_drugbank_drug_protein[['UniProt ID', 'DrugBank ID']].drop_duplicates()
drugbank_drug_protein
```

### BindingDB

```python
# Download Link: https://www.bindingdb.org/rwd/bind/downloads/BindingDB_All_202503_tsv.zip
# Download Date: 2025-03-21
# Download Version: 2025-03-01

filtered_data = []

chunksize = 10**5
for chunk in pd.read_csv('BindingDB_All.tsv', sep='\t', chunksize=chunksize, on_bad_lines='skip'):
    filtered_chunk = chunk[chunk['Target Source Organism According to Curator or DataSource'] == 'Homo sapiens']
    filtered_data.append(filtered_chunk)

df_bindingdb_human = pd.concat(filtered_data, ignore_index=True)
df_bindingdb_human
```

```python
df_bindingdb_filter = df_bindingdb_human[['PubChem CID', 'UniProt (SwissProt) Primary ID of Target Chain']]
df_bindingdb_filter = df_bindingdb_filter.dropna(subset=['UniProt (SwissProt) Primary ID of Target Chain'])
df_bindingdb_filter = df_bindingdb_filter.dropna(subset=['PubChem CID'])
df_bindingdb_filter.rename(columns={'PubChem CID': 'PubChem_CID', 'UniProt (SwissProt) Primary ID of Target Chain': 'UniProt_ID'}, inplace=True)
df_bindingdb_filter['PubChem_CID'] = df_bindingdb_filter['PubChem_CID'].astype(str)
df_bindingdb_filter['PubChem_CID'] = df_bindingdb_filter['PubChem_CID'].replace('\.0', '', regex=True)
df_bindingdb_filter.drop_duplicates(inplace=True)
df_bindingdb_filter
```

### DrugCentral

```python
# Download Link: https://unmtid-dbs.net/download/DrugCentral/2021_09_01/drug.target.interaction.tsv.gz
# Download Date: 2025-03-21
# Download Version: 2025-01-03

df_drugcentral = pd.read_csv('drug.target.interaction.tsv', sep='\t')
df_drugcentral = df_drugcentral[df_drugcentral['ORGANISM'] == 'Homo sapiens']
df_drugcentral = df_drugcentral[['STRUCT_ID', 'ACCESSION']]
df_drugcentral = df_drugcentral.rename(columns = {'STRUCT_ID': 'struct_id', 'ACCESSION': 'Uniprot'})
df_drugcentral
```

```python
# Download Link: SQL filter
# Download Date: 2024-10-7
# Download Version: 2024-10-7

drugcentral_id = pd.read_csv('drugcentral_drug_id_filter.csv')
drugcentral_id.head()
```

```python
drugcentral_drug_protein = pd.merge(df_drugcentral, drugcentral_id, how = 'left', on = 'struct_id')
drugcentral_drug_protein = drugcentral_drug_protein.drop(columns = ['struct_id'])
drugcentral_drug_protein
```

### BioMedgraphica ID

```python
import pandas as pd
import os
from pathlib import Path

current_working_dir = Path(os.getcwd()).resolve()
grandparent_dir = current_working_dir.parent.parent.parent
target_dir_drug = grandparent_dir / 'BioMedGraphica' / 'Entity' / 'Drug' / 'BioMedGraphica_Drug.csv'
target_dir_protein = grandparent_dir / 'BioMedGraphica' / 'Entity' / 'Protein' / 'BioMedGraphica_Protein.csv'
biomedgraphica_protein = pd.read_csv(target_dir_protein, dtype=str)
biomedgraphica_drug = pd.read_csv(target_dir_drug, dtype=str)
```

### DrugBank Mapping

```python
uniprot_id = biomedgraphica_protein[['Uniprot_ID', 'BioMedGraphica_ID']]
uniprot_id.dropna(subset = ['Uniprot_ID'], inplace=True)
uniprot_id = uniprot_id.assign(uniprot_id=uniprot_id['Uniprot_ID'].str.split(';')).explode('Uniprot_ID')

uniprot_protein_to_individualid = biomedgraphica_protein.groupby('Uniprot_ID')['BioMedGraphica_ID'].apply(lambda x: ';'.join(x.dropna().unique())).to_dict()
```

```python
drugbank_drug_id = biomedgraphica_drug[['DrugBank_ID', 'BioMedGraphica_ID']]
drugbank_drug_id.dropna(subset = ['DrugBank_ID'], inplace=True)
drugbank_drug_id = drugbank_drug_id.assign(drugbank_id=drugbank_drug_id['DrugBank_ID'].str.split(';')).explode('DrugBank_ID')

drugbank_drug_to_individualid = biomedgraphica_drug.groupby('DrugBank_ID')['BioMedGraphica_ID'].apply(lambda x: ';'.join(x.dropna().unique())).to_dict()
```

```python
drugbank_drug_protein['To_ID'] = drugbank_drug_protein['UniProt ID'].map(uniprot_protein_to_individualid)
drugbank_drug_protein['From_ID'] = drugbank_drug_protein['DrugBank ID'].map(drugbank_drug_to_individualid)
drugbank_drug_protein
```

```python
drugbank_d_p = drugbank_drug_protein[['From_ID', 'To_ID']]
drugbank_d_p.dropna(subset=['To_ID'], inplace=True)
drugbank_d_p.dropna(subset=['From_ID'], inplace=True)
drugbank_d_p
```

```python
drugbank_d_p['To_ID'] = drugbank_d_p['To_ID'].str.split(';')
drugbank_d_p = drugbank_d_p.explode('To_ID')

drugbank_d_p['From_ID'] = drugbank_d_p['From_ID'].str.split(';')
drugbank_d_p = drugbank_d_p.explode('From_ID')
drugbank_d_p.drop_duplicates(inplace=True)
drugbank_d_p
```

### BindingDB Mapping

```python
df_bindingdb_filter['To_ID'] = df_bindingdb_filter['UniProt_ID'].map(uniprot_protein_to_individualid)
df_bindingdb_filter
```

```python
CID_to_individualID = biomedgraphica_drug[['PubChem_CID', 'BioMedGraphica_ID']]
CID_to_individualID.dropna(subset = ['PubChem_CID'], inplace=True)
CID_to_individualID = CID_to_individualID.assign(PubChem_CID=CID_to_individualID['PubChem_CID'].str.split(';')).explode('PubChem_CID')
CID_to_individualID = CID_to_individualID.groupby('PubChem_CID')['BioMedGraphica_ID'].apply(lambda x: ';'.join(x.dropna().unique())).to_dict()

df_bindingdb_filter['From_ID'] = df_bindingdb_filter['PubChem_CID'].map(CID_to_individualID)
df_bindingdb_filter
```

```python
bindingdb_drug_protein = df_bindingdb_filter[['From_ID', 'To_ID']]
bindingdb_drug_protein.dropna(subset=['From_ID'], inplace=True)
bindingdb_drug_protein.dropna(subset=['To_ID'], inplace=True)

bindingdb_drug_protein
```

```python
bindingdb_drug_protein['To_ID'] = bindingdb_drug_protein['To_ID'].str.split(';')
bindingdb_drug_protein = bindingdb_drug_protein.explode('To_ID')

bindingdb_drug_protein['From_ID'] = bindingdb_drug_protein['From_ID'].str.split(';')
bindingdb_drug_protein = bindingdb_drug_protein.explode('From_ID')

bindingdb_drug_protein.drop_duplicates(inplace=True)
bindingdb_drug_protein.reset_index(drop=True, inplace=True)
bindingdb_drug_protein
```

### DrugCentral Mapping

```python
drugcentral_drug_protein['To_ID'] = drugcentral_drug_protein['Uniprot'].map(uniprot_protein_to_individualid)
drugcentral_drug_protein['Drug_drugbank'] = drugcentral_drug_protein['DRUGBANK_ID'].map(drugbank_drug_to_individualid)
drugcentral_drug_protein['Drug_cid'] = drugcentral_drug_protein['PUBCHEM_CID'].map(CID_to_individualID)
drugcentral_drug_protein
```

```python
def merge_string_columns(df, columns, merge_name, separator=';'):
    def merge_strings(row):
        combined = set()
        for column in columns:
            if pd.notnull(row[column]):
                combined.update(row[column].split(separator))
        return separator.join(combined)
    
    # Apply the function to each row and create a new column
    combined_column_name = merge_name
    df[combined_column_name] = df.apply(merge_strings, axis=1)
    df.drop(columns=columns, inplace=True)
    
    return df

drugcentral_drug_protein = merge_string_columns(drugcentral_drug_protein, ['Drug_drugbank', 'Drug_cid'], 'From_ID')
drugcentral_drug_protein
```

```python
drugcentral_drug_protein.drop(columns=['CHEBI', 'DRUGBANK_ID', 'PUBCHEM_CID', 'Uniprot'], inplace=True)
drugcentral_drug_protein.replace('', pd.NA, inplace=True)
drugcentral_drug_protein.dropna(subset=['To_ID'], inplace=True)
drugcentral_drug_protein.dropna(subset=['From_ID'], inplace=True)

drugcentral_drug_protein
```

```python
drugcentral_drug_protein['To_ID'] = drugcentral_drug_protein['To_ID'].str.split(';')
drugcentral_drug_protein = drugcentral_drug_protein.explode('To_ID')

drugcentral_drug_protein['From_ID'] = drugcentral_drug_protein['From_ID'].str.split(';')
drugcentral_drug_protein = drugcentral_drug_protein.explode('From_ID')

drugcentral_drug_protein.drop_duplicates(inplace=True)
drugcentral_drug_protein.reset_index(drop=True, inplace=True)
drugcentral_drug_protein
```

### Drug-Protein Relation

```python
drugbank_d_p['source1'] = 'DrugBank'
bindingdb_drug_protein['source2'] = 'BindingDB'
drugcentral_drug_protein['source3'] = 'DrugCentral'

drug_protein = pd.merge(drugbank_d_p, bindingdb_drug_protein, on=['From_ID', 'To_ID'], how='outer')
drug_protein = pd.merge(drug_protein, drugcentral_drug_protein, on=['From_ID', 'To_ID'], how='outer')
drug_protein
```

```python
def merge_string_columns(df, columns, merge_name, separator=';'):
    def merge_strings(row):
        combined = set()
        for column in columns:
            if pd.notnull(row[column]):
                combined.update(row[column].split(separator))
        return separator.join(combined)
    
    # Apply the function to each row and create a new column
    combined_column_name = merge_name
    df[combined_column_name] = df.apply(merge_strings, axis=1)
    df.drop(columns=columns, inplace=True)
    
    return df

drug_protein = merge_string_columns(drug_protein, ['source1', 'source2','source3'], 'Source')
drug_protein
```

```python
drug_protein['Type'] = 'Drug-Protein'

max_length = len(str(len(drug_protein)))
drug_protein['BioMedGraphica_ID'] = ['BMG_ED_DGPT' + str(i).zfill(max_length) for i in range(1, len(drug_protein) + 1)]
columns = ['BioMedGraphica_ID'] + [col for col in drug_protein.columns if col != 'BioMedGraphica_ID']  # re-order columns
drug_protein = drug_protein[columns]
drug_protein
```

```python
import os
from pathlib import Path

# get the current working directory
current_working_dir = Path(os.getcwd()).resolve()

# get the output directory
grandparent_dir = current_working_dir.parent.parent.parent

target_folder = grandparent_dir / 'BioMedGraphica' / 'Relation' / 'Drug-Protein'
if not target_folder.exists():
    target_folder.mkdir(parents=True)
    print(f"Folder {target_folder} has been created.")

output_file_path = target_folder / 'BioMedGraphica_Drug_Protein.csv'
drug_protein.to_csv(output_file_path, index=False)
print(f"Data has been saved to {output_file_path}")
```
