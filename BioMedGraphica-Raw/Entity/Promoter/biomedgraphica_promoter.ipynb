### File Generation

```python
import pandas as pd
import os
from pathlib import Path

current_working_dir = Path(os.getcwd()).resolve()
grandparent_dir = current_working_dir.parent.parent.parent
target_dir = grandparent_dir / 'BioMedGraphica' / 'Entity' / 'Promoter' / 'BioMedGraphica_Promoter.csv'
biomedgraphica_promoter = pd.read_csv(target_dir, dtype=str)
biomedgraphica_promoter
```

```python
def add_promoter_to_symbols_and_names(df, symbol_col, name_col):
    # Function to add " Promoter" to each element in the column
    def add_promoter(value):
        if pd.isna(value):
            return value
        # Split the value by ';' if there are multiple elements
        elements = value.split(';')
        # Add " Promoter" to each element
        return ';'.join([element.strip() + ' Promoter' for element in elements])

    # Apply the function to the columns if they exist
    if symbol_col in df.columns:
        df[symbol_col] = df[symbol_col].apply(add_promoter)
    if name_col in df.columns:
        df[name_col] = df[name_col].apply(add_promoter)
    
    return df


bmg_promoter = add_promoter_to_symbols_and_names(biomedgraphica_promoter, 'HGNC_Symbol', 'Gene_Name')
bmg_promoter
```

GUI Name

```python
def merge_string_columns(df, columns, merge_name, separator=' | '):
    def merge_strings(row):
        combined = set()
        for column in columns:
            if pd.notnull(row[column]):
                combined.update(row[column].split(separator))
        return separator.join(combined)
    
    # Apply the function to each row and create a new column
    combined_column_name = merge_name
    df[combined_column_name] = df.apply(merge_strings, axis=1)
    df.drop(columns=columns, inplace=True)
    
    return df

gui_name = bmg_promoter.copy()
gui_name = merge_string_columns(gui_name, ['Gene_Name', 'HGNC_Symbol'], 'Promoter_Name_List')
gui_name = gui_name[['BioMedGraphica_ID', 'Promoter_Name_List']]
gui_name
```

```python
import os
from pathlib import Path

# get the current working directory
current_working_dir = Path(os.getcwd()).resolve()

# get the output directory
grandparent_dir = current_working_dir.parent.parent.parent

target_folder = grandparent_dir / 'BioMedGraphica' / 'Entity' / 'Promoter'
if not target_folder.exists():
    target_folder.mkdir(parents=True)
    print(f"Folder {target_folder} has been created.")

output_file_path = target_folder / 'BioMedGraphica_Promoter_GUI_Name.csv'
gui_name.to_csv(output_file_path, index=False)
print(f"Data has been saved to {output_file_path}")
```

LLM Name and ID

```python
llm_name_id = bmg_promoter.copy()
llm_name_id.drop(columns=['Gene_Start', 'Gene_End', 'Chromosome', 'Gene_Type'], inplace=True)

llm_name_id['NCBI_Gene_ID'] = llm_name_id['NCBI_Gene_ID'].apply(
    lambda x: ' | '.join(f"NCBI Gene ID:{id}" for id in x.split(';')) if pd.notna(x) and x != '' else x
)
llm_name_id['RefSeq_ID'] = llm_name_id['RefSeq_ID'].apply(
    lambda x: ' | '.join(f"RefSeq ID:{id}" for id in x.split(';')) if pd.notna(x) and x != '' else x
)
llm_name_id['OMIM_ID'] = llm_name_id['OMIM_ID'].apply(
    lambda x: ' | '.join(f"OMIM ID:{id}" for id in x.split(';')) if pd.notna(x) and x != '' else x
)

column_order = ['BioMedGraphica_ID', 'HGNC_Symbol', 'Gene_Name', 
                'NCBI_Gene_ID', 'Ensembl_Gene_ID', 'HGNC_ID', 'RefSeq_ID', 'OMIM_ID']

llm_name_id = llm_name_id[column_order]
llm_name_id
```

```python
import os
from pathlib import Path

# get the current working directory
current_working_dir = Path(os.getcwd()).resolve()

# get the output directory
grandparent_dir = current_working_dir.parent.parent.parent

target_folder = grandparent_dir / 'BioMedGraphica' / 'Entity' / 'Promoter'
if not target_folder.exists():
    target_folder.mkdir(parents=True)
    print(f"Folder {target_folder} has been created.")

output_file_path = target_folder / 'BioMedGraphica_Promoter_LLM_Name_ID.csv'
llm_name_id.to_csv(output_file_path, index=False)
print(f"Data has been saved to {output_file_path}")
```

LLM Name and ID Combined

```python
llm_combined = llm_name_id.copy()

def merge_string_columns(df, columns, merge_name, separator=' | '):
    def merge_strings(row):
        combined = set()
        for column in columns:
            if pd.notnull(row[column]):
                combined.update(row[column].split(separator))
        return separator.join(combined)
    
    # Apply the function to each row and create a new column
    combined_column_name = merge_name
    df[combined_column_name] = df.apply(merge_strings, axis=1)
    df.drop(columns=columns, inplace=True)
    
    return df

llm_combined = merge_string_columns(llm_combined, ['HGNC_Symbol', 'Gene_Name', 'NCBI_Gene_ID', 
                'Ensembl_Gene_ID', 'HGNC_ID', 'RefSeq_ID', 'OMIM_ID'], 'Names_and_IDs')
llm_combined
```
